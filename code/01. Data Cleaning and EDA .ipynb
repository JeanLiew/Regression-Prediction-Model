## 5.0 EDA and Feature Engineering

### 5.1 Categorical Integer Encoding

There are multiple features that shares the same parameter of evaluation. In the dateaset, there are 23 ordinal features. Since the ordinal parameters has a logical ranking, the categorical features will be updated to rankings in integer where 0 is not applicable and onwards to represent a better score. 

```python
#rating updated to numerical ranking 23 ordinal cols

cols1 = ['exter_qual','exter_cond','bsmt_qual','bsmt_cond','heating_qc','kitchen_qual','fireplace_qu','garage_qual','garage_cond','pool_qc']
train[cols1] = train[cols1].replace({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0})

cols2 = ['bsmtfin_type_1','bsmtfin_type_2']
train[cols2] = train[cols2].replace({'GLQ': 6, 'ALQ': 5,'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0})

train['bsmt_exposure'] = train['bsmt_exposure'].replace({'Gd':4, 'Av':3, 'Mn':2, 'No':1, 'NA':0})

train['functional'] = train['functional'].replace({'Typ': 6, 'Min1': 5, 'Min2': 5,'Mod':4, 'Maj1':3, 'Maj2':2, 'Sev':1, 'Sal':0})

train['garage_finish'] = train['garage_finish'].replace({'Fin':3, 'RFn':2, 'Unf':1, 'NA':0})

train['lot_shape'] =  train['lot_shape'].replace({'Reg':4, 'IR1':3, 'IR2':2, 'IR3':1})

train['utilities'] =  train['utilities'].replace({'AllPub':4, 'NoSewr':3, 'NoSeWa':2, 'ELO':1})

train['land_slope'] =  train['land_slope'].replace({'Gtl':3, 'Mod':2, 'Sev':1})

train['bsmt_exposure'] = train['bsmt_exposure'].replace({'Gd':4, 'Av':3, 'Mn':2, 'No':1, 'NA':0})

train['garage_finish'] = train['garage_finish'].replace({'Fin':3, 'RFn':2, 'Unf':1, 'NA':0})

train['fence'] = train['fence'].replace({'GdPrv':4, 'MnPrv':3, 'GdWo':2, 'MnWw':1, 'NA':0})

#mixed to be classified equivalent to SBrkr
train['electrical'] = train['electrical'].replace({'SBrkr':4, 'FuseA':3, 'FuseF':2, 'FuseP':1, 'Mix':4})

train['paved_drive'] = train['paved_drive'].replace({'Y':3, 'P':2, 'N':1})
```

```python
#ordinal electrical to evaluate the Mix importance
train['electrical'].value_counts()
```

Mixed parameter in the elecrical feature is a stand-alone criteria which does not equate to better or worst. Since it has only 1 count, and a functional parameter, the ranking will be equivalent to the highest scoreing SBrKr (Standard Circuit Breaker & Romex)

```python
train.sample(5)
```

### 5.2 Numerical Feature Engineering

Instead of taking account to the year when it was built / modified, the age would be a relative and a tangible comparison for buyer as a prejudgment criteria on the house price. The general perception is that older houses may have more hidden issues and/or history. This would make better representation on its effect on the selling price.

Square feet measurements in the dataset have been divided to three parts, which total the square feet together as a summation would be a simplified manner to recognize the effect of housing space to the price.

```python
#create ages col to replace year built and year remod and garage year built from its year sold

train['age_built'] = train.yr_sold - train['year_built']
train['age_remod'] = train.yr_sold - train['year_remod/add']
train['age_garage'] = train.yr_sold - train['garage_yr_blt']
```

```python
#total sf includes 1st flr 2nd floor and bsmt
train['total_sf'] = train['1st_flr_sf'] + train['2nd_flr_sf'] + train['total_bsmt_sf']
```
