```python
#load training data set
#keep default param of NA, and empty string to nan

hold_out = pd.read_csv('../data/test.csv',na_values='',keep_default_na=False)
final = pd.read_csv('../cleaned_data/final_train_data.csv')
y = pd.read_csv('../cleaned_data/final_train_saleprice.csv')
```

```python
#check for mas_vnr type and area nan value
hold_out[hold_out.isnull().mas_vnr_area][['mas_vnr_type','mas_vnr_area']]
```

Mode for mas_vnr_type imputed in training set is 'None, and mas_vnr_area is 0. The same will be applied here. 

```python
#vnr area is 0 , mas vnr type is None
hold_out.loc[hold_out.mas_vnr_type.isnull(), 'mas_vnr_type']= 'None'
hold_out.loc[hold_out.mas_vnr_area.isnull(), 'mas_vnr_area']= 0
```

```python
hold_out.isnull().sum().sort_values().tail()
```

```python
#check balance 2 missing value and dropped
hold_out[hold_out.isnull().sum(axis=1)>0]
```

Since there is 4 missing rows with nan values of the 2051 entries, these 4 rows will be dropped.

### 1.2 Feature Engineering

```python
#rating updated to numerical ranking 23 ordinal cols

cols1 = ['exter_qual','exter_cond','bsmt_qual','bsmt_cond','heating_qc','kitchen_qual','fireplace_qu','garage_qual','garage_cond','pool_qc']
hold_out[cols1] = hold_out[cols1].replace({'Ex':5, 'Gd':4, 'TA':3, 'Fa':2, 'Po':1, 'NA':0})

cols2 = ['bsmtfin_type_1','bsmtfin_type_2']
hold_out[cols2] = hold_out[cols2].replace({'GLQ': 6, 'ALQ': 5,'BLQ':4, 'Rec':3, 'LwQ':2, 'Unf':1, 'NA':0})

hold_out['bsmt_exposure'] = hold_out['bsmt_exposure'].replace({'Gd':4, 'Av':3, 'Mn':2, 'No':1, 'NA':0})

hold_out['functional'] = hold_out['functional'].replace({'Typ': 6, 'Min1': 5, 'Min2': 5,'Mod':4, 'Maj1':3, 'Maj2':2, 'Sev':1, 'Sal':0})

hold_out['lot_shape'] =  hold_out['lot_shape'].replace({'Reg':4, 'IR1':3, 'IR2':2, 'IR3':1})

hold_out['utilities'] =  hold_out['utilities'].replace({'AllPub':4, 'NoSewr':3, 'NoSeWa':2, 'ELO':1})

hold_out['land_slope'] =  hold_out['land_slope'].replace({'Gtl':3, 'Mod':2, 'Sev':1})

hold_out['bsmt_exposure'] = hold_out['bsmt_exposure'].replace({'Gd':4, 'Av':3, 'Mn':2, 'No':1, 'NA':0})

hold_out['fence'] = hold_out['fence'].replace({'GdPrv':4, 'MnPrv':3, 'GdWo':2, 'MnWw':1, 'NA':0})

hold_out['paved_drive'] = hold_out['paved_drive'].replace({'Y':3, 'P':2, 'N':1})
```

```python
#create ages col to replace year built and year remod and garage year built from its year sold

hold_out['age_built'] = hold_out.yr_sold - hold_out['year_built']
hold_out['age_remod'] = hold_out.yr_sold - hold_out['year_remod/add']
hold_out['age_garage'] = hold_out.yr_sold - hold_out['garage_yr_blt']
```

```python
#total_sf is to include 1st flr, 2nd flr and basement sf
hold_out['total_sf'] = hold_out['1st_flr_sf'] + hold_out['2nd_flr_sf'] + hold_out['total_bsmt_sf']
```

```python
hold_out.loc[hold_out.ms_zoning.isin(['C (all)', 'R (all)','I (all)', 'A (agr)','FV']), 'ms_zoning'] = 'Others'
```

```python
hold_out.ms_zoning.unique()
```

```python
#inside corner FR2 in the same grouping 

hold_out['open_lot'] = hold_out['lot_config'].isin(['FR3'])
hold_out['common_lot'] = hold_out['lot_config'].isin(['Corner', 'Inside', 'FR2'])
hold_out['closed_lot'] = hold_out['lot_config'].isin(['CulDSac'])
```

```python
#create col for PosA and PosN based on condition 2
#create col to group other together and normal is excluded

hold_out['near_park'] = hold_out['condition_2'].isin(['PosA', 'PosN'])
hold_out['near_road'] = hold_out['condition_2'].isin(['RRNn', 'Feedr', 'Artery', 'RRAe', 'RRAn'])
```

```python
#function to change to one-hot encode categorical object
cat_df = hold_out.select_dtypes(exclude=['number','bool'])

def one_hot(dataframe):
    for i in dataframe:
        dataframe = pd.get_dummies(dataframe, drop_first=False)
        return dataframe

#merge get-dummied data in original dat set
X = pd.concat([hold_out,one_hot(cat_df)],axis=1)
```

```python
X.columns
```

```python
X['heating_OthW'] = 0
```

```python
X = X[X[final_cols].columns]
```

```python
X.head()
```

## 2.0 Model Prediction with Linear Regression

```python
final.head()
```

```python
#instantiate model

lr = LinearRegression()
lr.fit(final,y)
lr.score(final,y)
```

```python
X['SalePrice'] = lr.predict(X)
X['Id'] = hold_out.id
```

```python
submission = X[['Id', 'SalePrice']].sort_values('Id')
submission.head()
```

## 3.0 Export Prediction to CSV

```python
submission.to_csv('../cleaned_data/submission_kaggle.csv', index=False)
```
